```python
from wdlen_Thal import wordlenth
from nltk.corpus.reader import CHILDESCorpusReader
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
corpus_root = nltk.data.find( 'corpora/CHILDES/Spanish/')
Thal = CHILDESCorpusReader(corpus_root, 'JacksonThal/.*.xml')
wdlen = wordlenth(Thal)
```

```python
df = pd.DataFrame(wdlen).dropna()
df = df[['fname', 'age', 'one','two','three', 'four', 'five']]
df
```

```python
writer = pd.ExcelWriter('check spanish.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```

```python
plt.figure()
df2 = df.groupby('age').mean().reset_index()
df2.set_index('age').plot.line(rot=0, title='Population', figsize=(15,10), fontsize=12)
plt.show()
```
